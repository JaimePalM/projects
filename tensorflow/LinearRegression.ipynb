```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
height = np.array([5.2, 5.7, 5.9, 6.0, 6.4])
weight = np.array([125, 140, 155, 160, 180])
```

```python
plt.scatter(height,weight)
plt.xlabel('Height (inches)')
plt.ylabel('Weight (pounds)')
plt.show()
```

```python
m, b = np.polyfit(height, weight, 1)
```

```python
plt.scatter(height,weight)
plt.plot(height, m * height + b, color = 'red')
plt.xlabel('Height (inches)')
plt.ylabel('Weight (pounds)')
plt.show()
```

```python
new_height = 5.8
predicted_weight = m* new_height + b
print('Predicted Weight: ', predicted_weight)
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('car_purchasing_data.csv')
data
```

```python
X = data[['gender','age','salary','debt','worth']]
y = data['amount']
```

```python
model = LinearRegression()
model.fit(X,y)
```

```python
new_data = np.array([[1, 33, 75500, 15645, 335845]])
predicted_price = model.predict(new_data)
print(f'Purchasing price would be {predicted_price}')
```
