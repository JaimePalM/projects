```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans 
```

```python
#Generate random data points
X,_= make_blobs(n_samples = 300, centers = 4, random_state = 42)
```

```python
#plot the data points
plt.scatter(X[:,0],X[:,1], s= 50)
plt.xlabel("feature 1")
plt.ylabel("feature 2")
plt.title("Data Points")
plt.show()
```

```python
# create a k-means oject with k=4 clusters
kmeans = KMeans(n_clusters = 4, random_state = 42)

#Fit the K-means to the data
kmeans.fit(X)
```

```python
#Get the cluster labels assigned to each data points
labels = kmeans.labels_

#Get the cluster centers
centers= kmeans.cluster_centers_

#plot the data points with cluster asssignement
plt.scatter(X[:,0],X[:,1], c=labels,cmap = "viridis")

# plot the cluster centers

plt.scatter(centers[:,0],  centers[:,1], marker = "o", c="red", s=200, alpha=0.7)

plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("K-Means Clustering")
plt.show()
```
