```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
```

```python
#load the dataset

data = load_breast_cancer()
```

```python
X = data.data
y = data.target
```

```python
# split the dtaa into training and testing set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size =0.3 , random_state = 42)
```

```python
# using 100 decision trees, random is to ensure reproductibility
rf_classifier = RandomForestClassifier(n_estimators = 100, random_state= 42)
```

```python
#train the classifier
rf_classifier.fit(X_train, y_train)
```

```python
# Make predictions on the test set
predictions = rf_classifier.predict(X_test)
```

```python
# evaluate the classifier
import sklearn
accuracy =sklearn.metrics.accuracy_score(y_test,predictions)
print(f"Accuracy:{accuracy}")
```

```python
# get feature importances
importances =  rf_classifier.feature_importances_
```

```python
#sort feature importances indescneding order
indices = np.argsort(importances[::-1])

#rearrange features names do they match the sorted feature importances
feature_names = data.feature_names[indices]
```

```python
# create bar plot of feature importances

plt.figure(figsize = (10,6))
plt.title("feature Importances")
plt.bar(range(X.shape[1]),importances[indices])
plt.xticks(range(X.shape[1]),feature_names, rotation = 90)
plt.show()
```
